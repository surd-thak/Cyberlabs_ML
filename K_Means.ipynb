```python
def Kmeans_model(DF1, k, s):
  Int = [Int for Int in range(0,14800)]
  centroid = DF1[np.random.choice(Int, k, replace = False)]  
  for t in range(s):
    euc_list = []
  
    for i in range(k):
     euc_dist = 0
     euc_dist = np.sqrt(np.sum(((DF1 - centroid[i])**2), axis=1)) 
     euc_list.append(euc_dist)

    closest = np.argmin(euc_list, axis = 0) 
    centroidu = np.empty_like(centroid) 
    for j in range(k):
      arr = DF1[np.where(closest == j)]
      cen = 0
      cen = np.mean(arr, axis =0)
      centroidu[j] = cen
    
    if centroid.all == centroidu.all:
      break
    
    
    print(closest)
    
  centroid = centroidu
  return closest  
```
